```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
dias = np.arange(4, 16)
```

```python

print(dias)
```

```python
casos = [1, 2, 2, 8, 8, 11, 17, 19, 31, 34, 45, 56]
```

```python
muertes = [6, 10]
ticks = []
for i in range(4, 16):
    ticks.append("{}/3".format(i))
ticks
```

```python
fig = plt.figure()

plt.plot(dias, casos, ".-", label = "Casos de infección detectada")
plt.xticks(dias, ticks, rotation = "50")
plt.vlines(muertes, 0, [2, 17], linestyle = "dashed")
plt.hlines([2, 17], 4, muertes, linestyle = "dashed")
plt.annotate("Primera muerte", (6, 2), xytext = (3.8, 10), size = 8, arrowprops=dict(arrowstyle="->"))
plt.annotate("Segunda muerte", (10, 17), xytext = (7.5, 30),  size = 8, arrowprops=dict(arrowstyle="->"))
fit = np.polyfit(dias, np.log(casos), 1)
fity = np.exp(fit[1])*np.exp(fit[0]*dias)
plt.plot(dias, fity, "--", label = "Curva exponencial")
plt.title("Evolución Coronavirus Argentina")
plt.legend()
```
