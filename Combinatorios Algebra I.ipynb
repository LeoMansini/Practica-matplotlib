```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.figure as fig
```

```python
def factorial(x):
    a = 1
    if type(x) == int:
        for i in range(x):
            a = a * (i+1)
    return a
```

```python
print(factorial(8))
```

```python
def combinatorio(n, k):#usa funcion factorial y explica EN INGLISH
    if type(n) == int and type(k) == int:
        if n>=k:
            a = int(factorial(n)/(factorial(n-k)*factorial(k)))
            b = 'The amount of subsets of a set with ' + str(n) + ' elements that contain '+ str(k) + ' elements is ' + str(a) + '.'
            print(b)
        else:
            print("Second input is smaller than first.")
            
    else:
        print("Need two integers as input.")
```

```python
def combinatorio2(n, k):#usa funcion factorial devuelve solo el numero
    if type(n) == int and type(k) == int:
        if n>=k:
            a = int(factorial(n)/(factorial(n-k)*factorial(k)))
            return(a)
        else:
            print("Second input is smaller than first.")
            
    else:
        print("Need two integers as input.")
```

```python
combinatorio(5, 3)
```

```python
combinatorio(3, 2.2)
```

```python
f, axes = plt.subplots(nrows=2, ncols=3, figsize=(12,8), constrained_layout=True)
f.suptitle("Simetria de Combinatorios", y = 1.04, size = "20")
n = 4
for k in range(1, 3):
    for j in range(1, 4):# Los hago asi a los range para no confundir el numero de fila/columna.
        lin = range(1, n)
        comb = []
        for i in lin:
            comb.append(combinatorio2(n, i))
        comb = np.array(comb)
        axes[k-1,j-1].plot(lin, comb, ".r-", linewidth = "0.7", markersize = "10")
        axes[k-1,j-1].set_title(label = r"Para n = {} y k = (1, {})".format(n, n-1), fontsize = 15)
        axes[k-1,j-1].grid()
        n = n+1 # Sigo con el siguiente gráfico.
plt.savefig(r"C:\Users\LEO\Desktop\FCEN\cs\Algebra\Simetria de combinatorios.png", dpi = 800, bbox_inches='tight')
```
